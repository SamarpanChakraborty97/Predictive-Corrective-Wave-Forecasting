```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"]="serif"
plt.rcParams["font.size"]=12
```

```python
import os
import pandas as pd

num_layers = [1,2,4,5,8]
mae_errors = []
mse_errors = []
corr = []

for i in range(len(num_layers)):
    filename = os.getcwd() + "/" + "num_layers_" + f"{num_layers[i]}.csv"
    errors = np.asarray(pd.read_csv(filename, header=None))
    mae = errors[0]
    mse = errors[1]
    r = errors[2]
    mae_errors.append(mae)
    mse_errors.append(mse)
    corr.append(r)

plt.figure(figsize=[8,6], dpi=200)
plt.plot(num_layers, mae_errors, linestyle = '-', color='b', marker = 's', markersize = 5, markerfacecolor = 'b', markeredgecolor = 'b', label = 'MAE (m)')
# plt.plot(num_layers, mse_errors, linestyle = '-', color='r', marker = 'h', markersize = 5, markerfacecolor = 'r', markeredgecolor = 'r', label = 'MSE')
plt.plot(num_layers, corr, linestyle = '-', color='k', marker = 'o', markersize = 5, markerfacecolor = 'k', markeredgecolor = 'k', label = 'r')
plt.grid(which='both', alpha=0.2)
plt.xlabel("Number of graph layers")
plt.ylabel("Error metrics")
plt.legend()
plt.savefig("Error comparisons with different number of graph layers.png")
plt.show()
```
